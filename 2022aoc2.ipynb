```python
# bring in relevant libraries
import pandas as pd
import numpy as np
from pathlib import Path

# expand to width of screen
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
# import file, check output
folder = Path("/Users/gflo247/Desktop/programming/AOC/2022")
filename = "2022aoc2.txt"
with open(folder / filename) as f:
    mylist = f.read().splitlines()
mylist[:20]
```

```python
# convert list to dataframe, add column name, check output
inp0 = pd.DataFrame(mylist)

# new data frame with split value columns, keep original non-split column
inp1 = inp0[0].str.split(" ", n = 1, expand = True)
inp1 = pd.concat([inp0, inp1], axis=1)

# name columns from new dataframe & check output
inp1.columns = ['both','opp','you']
inp1.head(20)
```

```python
# create letter to rock/paper/scissor translation
ltr_to_rps = {
    'A': 'rock',
    'B': 'paper',
    'C': 'scissors',
    'X': 'rock',
    'Y': 'paper',
    'Z': 'scissors'
}

# create choice translation
rps_to_pts = {
    'X': 1,
    'Y': 2,
    'Z': 3
}

# create result translation
res_to_pts = {
    'A X': 3,
    'A Y': 6,
    'A Z': 0,
    'B X': 0,
    'B Y': 3,
    'B Z': 6,
    'C X': 6,
    'C Y': 0,
    'C Z': 3
}
```

```python
# apply all the above dictionaries & check output
inp1['rps_opp'] = inp1['opp'].replace(ltr_to_rps)
inp1['rps_you'] = inp1['you'].replace(ltr_to_rps)
inp1['pts_chc'] = inp1['you'].replace(rps_to_pts)
inp1['pts_res'] = inp1['both'].replace(res_to_pts)
inp1['tot_pts'] = inp1['pts_chc'] + inp1['pts_res']
inp1.head(20)
```

```python
sum(inp1['tot_pts'])
```

```python
# create letter to outcome translation
ltr_to_out = {
    'X': 'lose',
    'Y': 'draw',
    'Z': 'win'
}

# create outcome translation
out_to_pts = {
    'X': 0,
    'Y': 3,
    'Z': 6
}

# create outcome translation
rps_to_pts2 = {
    'rock': 1,
    'paper': 2,
    'scissors': 3
}

# create total points translation
both_to_chc = {
    'A X': 'scissors',
    'A Y': 'rock',
    'A Z': 'paper',
    'B X': 'rock',
    'B Y': 'paper',
    'B Z': 'scissors',
    'C X': 'paper',
    'C Y': 'scissors',
    'C Z': 'rock'
}
```

```python
# apply all the above dictionaries & check output
inp1['outcome'] = inp1['you'].replace(ltr_to_out)
inp1['chc_new'] = inp1['both'].replace(both_to_chc)
inp1['pts_chc_new'] = inp1['chc_new'].replace(rps_to_pts2)
inp1['pts_out_new'] = inp1['you'].replace(out_to_pts)
inp1['tot_pts_new'] = inp1['pts_chc_new'] + inp1['pts_out_new']
inp1.head(20)
```
