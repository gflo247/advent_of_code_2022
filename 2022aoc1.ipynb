```python
# bring in relevant libraries
import pandas as pd
import numpy as np

# expand to width of screen
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
# import file from work library ensuring it retains blanks for easier grouping
with open("2022aoc1.txt") as f:
    mylist = f.read().splitlines()
mylist[:20]
```

```python
# convert list to dataframe, add column name, check output
inp = pd.DataFrame(mylist)
inp.columns=['cals']
inp.head(20)
```

```python
# create separate index column
initial_value = 0
inp['idx'] = range(initial_value, len(inp) + initial_value)
inp.head(20)
```

```python
inp.tail(20)
```

```python
# save rows that possess blanks
blnks = np.where(inp['cals'] == '')
blnks
```

```python
# add max index value to end of array
blnks = np.append(blnks, 2254)
```

```python
# check total count
np.count_nonzero(blnks)
```

```python
# add in elf look-up values
blnk_rows = pd.DataFrame(blnks, columns = ['idx'])
initial_value = 1
blnk_rows['elf'] = range(initial_value, len(blnk_rows) + initial_value)
blnk_rows
```

```python
# merge w/ vlookup true type of function
inp_w_elves = pd.merge_asof(inp, blnk_rows, on=['idx'], direction='forward')
inp_w_elves['cals'] = pd.to_numeric(inp_w_elves['cals'], errors='ignore')
inp_w_elves
```

```python
# check that merge worked
inp_w_elves.head(50)
```

```python
# sum by elf
tot_cals_by_elf = inp_w_elves.groupby(['elf']).agg({'cals':'sum'})
tot_cals_by_elf
```

```python
# pull max
max(tot_cals_by_elf['cals'])
```

```python
# check elf w/ max cals
elf_piv = pd.pivot_table(tot_cals_by_elf, index=['elf'], values=['cals'], aggfunc=np.sum).sort_values(by=['cals'], ascending=False)
elf_piv
```

```python
elf_tbl = pd.DataFrame(elf_piv.to_records())
top3 = elf_tbl[elf_tbl['cals'] > 68500]
top3
```
