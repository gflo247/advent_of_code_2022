```python
# bring in relevant libraries
import pandas as pd
import numpy as np
import string
from pathlib import Path

# expand to width of screen
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
# import file, check output
folder = Path("/Users/gflo247/Desktop/programming/AOC/2022")
filename = "2022aoc3.txt"
with open(folder / filename) as f:
    mylist = f.read().splitlines()
mylist[:20]
```

```python
# convert list to dataframe, add column name, check output
inp0 = pd.DataFrame(mylist)
inp0.columns=['ruck']
inp0
```

```python
# print relevant (half)lengths of each row in a separate columns, produce 1st and 2nd compartment outputs, and resultant similar items
# inp0['ruck_len'] = inp0['ruck'].apply(len)
inp0['half_ruck_len'] = inp0['ruck'].apply(len)/2
inp0['half_ruck_len'] = inp0['half_ruck_len'].astype(int)
inp0['comp1'] = inp0.apply(lambda inp0: inp0['ruck'][:inp0['half_ruck_len']], axis=1)
inp0['comp2'] = inp0.apply(lambda inp0: inp0['ruck'][inp0['half_ruck_len']:], axis=1)
inp0['sim'] = inp0.apply(lambda inp0: ''.join(sorted(set(inp0['comp1']) & set(inp0['comp2']), key = inp0['comp1'].index)), axis=1)
inp0
```

```python
# create lists to build dictionaries for priority translation, check dictionaries
lo_nums = [i for i in range(1,27,1)]
alph_lo = list(string.ascii_lowercase)
alph_up = list(string.ascii_uppercase)

lo_dic = {}
for i in range(len(alph_lo)):
    lo_dic[alph_lo[i]] = lo_nums[i]
print('dictionary_low:', lo_dic)

up_dic = {}
for i in range(len(alph_up)):
    up_dic[alph_up[i]] = lo_nums[i] + 26
print('dictionary_high:', up_dic)
```

```python
# apply above dictionaries & check output
inp0['priority'] = inp0['sim'].replace(lo_dic).replace(up_dic)
inp0
```

```python
sum(inp0['priority'])
```

```python
def create_chunks(l, n):
     
    # looping thru length l
    for i in range(0, len(l), n):
        # creating chunks of size n
        yield l[i:i + n]
        
# loop through list to divide it into multiple chunks of 3 items in preparation for 3 columns
new_list = list(create_chunks(mylist, 3))

# convert list to dataframe, add column name, check output
inp1 = pd.DataFrame(new_list)
inp1.columns=['ruck1','ruck2','ruck3']
inp1
```

```python
# pull the similar letter from each group's rucksack, 
inp1['grp_sim'] = inp1.apply(lambda inp1: ''.join(sorted(set(inp1['ruck1']) & set(inp1['ruck2']) & set(inp1['ruck3']), key = inp1['ruck1'].index)), axis=1)

# apply above dictionaries & check output
inp1['grp_priority'] = inp1['grp_sim'].replace(lo_dic).replace(up_dic)
inp1
```
